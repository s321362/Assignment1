```python
import pandas as panda
import matplotlib
from matplotlib import pyplot as plot
```

```python
dataframe = panda.read_csv("Ruter_data.csv", delimiter=";")
dataframe.head() #shows the 5 first rows of the data set to verify the file is correctly imported.
```

```python
dataframe.shape
```

```python
plot.hist(dataframe.Passasjerer_Ombord, bins=10, rwidth=2)
plot.xlabel('Passasjerer_Ombord')
plot.ylabel('Count')
plot.show()
```

```python
dataframe.Passasjerer_Ombord.describe()
```

```python
dataframe['Ledig_kapasitet'] = dataframe['Kjøretøy_Kapasitet'] - dataframe['Passasjerer_Ombord'] #gives us a actual free capasity for each TurID 
```

```python
dataframe.head()
```

```python
plot.hist(dataframe.Ledig_kapasitet, bins=10, rwidth=0.8)
plot.xlabel('Antall ledige plasser per TurID fordelt på mengder')
plot.ylabel('Count')
plot.show()
```

```python
dataframe['Kapasitetutnyttelse'] = dataframe['Passasjerer_Ombord'] / dataframe['Kjøretøy_Kapasitet'] * 100 #gives us an percentage value of capacity utilization for each TurID 
```

```python
dataframe.head()
```

```python
plot.hist(dataframe.Kapasitetutnyttelse, bins=10, rwidth=0.8)
plot.xlabel('Kapasitetsutnyttelse i %')
plot.ylabel('Count')
plot.show()
```
